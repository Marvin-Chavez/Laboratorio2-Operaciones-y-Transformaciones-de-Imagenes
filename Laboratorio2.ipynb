##### Problema 1: Desarrolle una función la cual reciba dos parámetros, una imagen y un entero llamado color, la función debe devolver una imagen la cual tenga activos los canales de color según los siguientes puntos: 
##### •	Si el parámetro color vale 1, la imagen debe mostrar activos únicamente el color azul. 
##### •	Si el parámetro color vale 2, la imagen debe mostrar activos únicamente el color verde. 
##### •	Si el parámetro color vale 3, la imagen debe mostrar activos únicamente el color rojo. 
##### •	Si el parámetro color vale 10, la imagen debe mostrar activos únicamente los colores rojo y verde.
##### •	Si el parámetro color vale 20, la imagen debe mostrar activos únicamente los colores verde y azul. 
##### •	Si el parámetro color vale 30, la imagen debe mostrar activos únicamente los colores azul y rojo.


```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def cambiar_canales(imagen, color):
    # Dimensiones de la imagen
    alto = imagen.shape[0]
    ancho = imagen.shape[1]

    imagen_resultante = np.zeros((alto, ancho, 3))  # Espacio para la imagen resultante.

    for i in range(alto):
        for j in range(ancho):
            pixel = imagen[i, j]
            # Extraer canales.
            blue = pixel[0]
            green = pixel[1]
            red = pixel[2]
            # Aplicar transformación según el valor de 'color'.
            if color == 1:  # Mostrar solo el color azul.
                imagen_resultante[i, j] = [blue, 0, 0]
            elif color == 2:  # Mostrar solo el color verde.
                imagen_resultante[i, j] = [0, green, 0]
            elif color == 3:  # Mostrar solo el color rojo.
                imagen_resultante[i, j] = [0, 0, red]
            elif color == 10:  # Mostrar rojo y verde.
                imagen_resultante[i, j] = [red, green, 0]
            elif color == 20:  # Mostrar verde y azul.
                imagen_resultante[i, j] = [0, green, blue]
            elif color == 30:  # Mostrar azul y rojo.
                imagen_resultante[i, j] = [red, 0, blue]

    return imagen_resultante
```

```python
# Cargar la imagen
img = cv2.imread("problema1.jpg")
# Cambiar formato de carga de BGR a RGB.
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Llamar a la función para cambiar los canales.
imagen_resultante = cambiar_canales(img, 2)
#escribir imagen convertida
cv2.imwrite("imagen_resultante.jpg", imagen_resultante) 
iresultante = cv2.imread("imagen_resultante.jpg") #leer y convertir a RGB imagen convertida
iresultante = cv2.cvtColor(iresultante, cv2.COLOR_BGR2RGB)

# Mostrar la imagen resultante
plt.imshow(iresultante)
plt.show()
```

##### Problema 2:
##### En el .zip del laboratorio se le compartio un conjunto de imagenes en escala de grises, estas imagenes
##### fueron creadas utilizando una escala de grises en 3D, cree una funcion que dadas las 3 imagenes se
##### construya la imagen original a color.

```python
def reconstruir_imagen_color(img_gris_rojo, img_gris_verde, img_gris_azul):
    # Cargamos las imágenes en escala de grises
    img_rojo = cv2.imread(img_gris_rojo, cv2.IMREAD_GRAYSCALE)
    img_verde = cv2.imread(img_gris_verde, cv2.IMREAD_GRAYSCALE)
    img_azul = cv2.imread(img_gris_azul, cv2.IMREAD_GRAYSCALE)

    # Obtenemos las dimensiones de las imágenes
    alto = img_rojo.shape[0]
    ancho = img_rojo.shape[1]

    # Creamos la imagen resultante a color
    img_color = np.zeros((alto, ancho, 3), dtype=np.uint8)

    # Asignamos los canales de color en la imagen resultante
    img_color[:, :, 0] = img_azul
    img_color[:, :, 1] = img_verde
    img_color[:, :, 2] = img_rojo

    return img_color

```

```python

img_gris_rojo = "imagen1_salida_gray_rojo.jpg"
img_gris_verde = "imagen1_salida_gray_verde.jpg"
img_gris_azul = "imagen1_salida_gray_azul.jpg"

imagen_original_color = reconstruir_imagen_color(img_gris_rojo, img_gris_verde, img_gris_azul)

#escribir imagen convertida
cv2.imwrite("imagen_original_color.jpg", imagen_original_color) 
icolor = cv2.imread("imagen_original_color.jpg") #leer y convertir a RGB imagen convertida
icolor = cv2.cvtColor(icolor, cv2.COLOR_BGR2RGB)

# Mostrar la imagen resultante
plt.imshow(icolor)
plt.show()
```

##### Problema 4: Cree una función que dada una imagen, muestre el histograma de cada canal de color y el de escala de grises (utilice un promedio aritmético para su escala de grises, no puede usar funciones de opencv), sus histogramas deben incluir una línea vertical la cual muestre el valor de la media de la distribución.

```python

def histograma(imagen):
    # Leer la imagen
    img_color = cv2.imread(imagen)
    alto, ancho, _ = img_color.shape

    #Calulando los histogramas
    hist_rojo = np.histogram(img_color[:, :, 2], bins=256, range=[0, 256])[0]
    hist_verde = np.histogram(img_color[:, :, 1], bins=256, range=[0, 256])[0]
    hist_azul = np.histogram(img_color[:, :, 0], bins=256, range=[0, 256])[0]

     #Calculando a mano los valores con for 
    img_gris = np.zeros((alto, ancho), dtype=np.uint8)
    for i in range(alto):
        for j in range(ancho):
            pixel = img_color[i, j]
            img_gris[i, j] = int((int(pixel[0]) + int(pixel[1]) + int(pixel[2])) / 3)

    hist_gris = np.histogram(img_gris, bins=256, range=[0, 256])[0]

    #La media
    media_rojo = np.sum(hist_rojo * np.arange(256)) / np.sum(hist_rojo)
    media_verde = np.sum(hist_verde * np.arange(256)) / np.sum(hist_verde)
    media_azul = np.sum(hist_azul * np.arange(256)) / np.sum(hist_azul)
    media_gris = np.sum(hist_gris * np.arange(256)) / np.sum(hist_gris)

    #Impresion de resultados
    plt.figure(figsize=(12, 6))
    plt.subplot(2, 2, 1)
    plt.plot(hist_rojo, color='red')
    plt.axvline(media_rojo, color='black', linestyle='dashed', linewidth=1)
    plt.title("Histograma Rojo")
    plt.subplot(2, 2, 2)
    plt.plot(hist_verde, color='green')
    plt.axvline(media_verde, color='black', linestyle='dashed', linewidth=1)
    plt.title("Histograma Verde")
    plt.subplot(2, 2, 3)
    plt.plot(hist_azul, color='blue')
    plt.axvline(media_azul, color='black', linestyle='dashed', linewidth=1)
    plt.title("Histograma Azul")
    plt.subplot(2, 2, 4)
    plt.plot(hist_gris, color='gray')
    plt.axvline(media_gris, color='black', linestyle='dashed', linewidth=1)
    plt.title("Histograma Escala de Grises")
    plt.tight_layout()
    plt.show()


```

```python

imagen = "Problema3.jpg"
histograma(imagen)
```
